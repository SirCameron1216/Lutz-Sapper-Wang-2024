```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import time

from sklearn.linear_model import LinearRegression

# Importing Package for RFE
from sklearn.feature_selection import RFE
```

```python
solvents_sapper = pd.read_csv('Chemopy_Descriptors_for_Cameron.csv')
solvents_sapper
```

```python
solvents_rhelogy = {
    'Density': [1105,789,1261,1090,868,970,995,830,750,814,816,867,935],      #units are in kg/m3
    'Viscosity': [15.78,1.1,934,27.09,0.81,1.8,1,12.1,1.3,4.3,7.4,0.560,97],    #units are in mPa*s
    'Surface Tension': [45.51,21.97,76.2,42.73,28.8,35.9,29.3,28.5,25.4,24.5,26.5,28.3,33]   #units are in mN/m at 25C
}
solvents_rhelogy_frame = pd.DataFrame(solvents_rhelogy)
solvents_rhelogy_frame
```

## Density

```python
density_df = solvents_rhelogy_frame['Density']
density_df

dens_desc = pd.DataFrame.join(solvents_sapper,density_df)
dens_desc
```

```python
dens_desc.head()
dens_desc_clean = dens_desc.drop(columns=['Molecule','SMILES'])
dens_desc_clean
```

```python
d_y = dens_desc_clean['Density']
d_X = dens_desc_clean.drop(columns=['Density'])
dtrain_X, dtest_X, dtrain_y, dtest_y = train_test_split(d_X, d_y, test_size = 0.2, random_state = 16)
```

```python
corr = d_X.corr()

# Step 2: Set correlations below 70% to NaN
# You might want to keep the diagonal (self-correlations) by using np.fill_diagonal after this step
corr_filtered = corr.where(np.abs(corr) >= 0.7)

# Optional: Mask out the upper triangle for a cleaner look
#mask = np.triu(np.ones_like(corr_filtered, dtype=bool))

# Step 3: Create the heatmap
plt.figure(figsize=(10, 8))  # You can adjust the figure size as needed
sns.heatmap(corr_filtered, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix (Only > 70%)')
plt.show()
```

```python
density_lm = LinearRegression() 

dens_rfe = RFE(estimator = density_lm, n_features_to_select = 50)

dens_rfe = dens_rfe.fit(dtrain_X, dtrain_y)
```

```python
drank = pd.DataFrame(list(zip(dtrain_X.columns, dens_rfe.support_, dens_rfe.ranking_)),
             columns = ['Feature','Whether selected', 'Rank'])

drank_sorted = pd.DataFrame.sort_values(drank,by=['Rank'],ascending=True)
drank_sorted
```

```python
# Predict Test Set
dy_pred = dens_rfe.predict(dtest_X)

# evaluate the model on test set
r2_density = sklearn.metrics.r2_score(dtest_y, dy_pred)
print('R-squared on Test Set: %0.2f' %r2_density)

RMSE_test_density = sklearn.metrics.mean_squared_error(dtest_y, dy_pred, squared=False)
print('RMSE on Test Set: %0.2f' %RMSE_test_density)
```

```python
df_true = drank.loc[drank['Whether selected'] == True]
df_true = df_true.drop(columns=['Whether selected'])
df_true["Coefficients"] = dens_rfe.estimator_.coef_

print('intercept ', dens_rfe.estimator_.intercept_)
df_true
```
